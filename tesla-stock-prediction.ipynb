```python
###This is a program for prediction of stock movement and price prediction  by using the machine learning algorithm
###The stock prediction or the price prediction is one of the latest trends in financial market because investments can be make safer by algo trading .
###The machine learning algorithm that will be used is randomforestregressor and the datset si taken from kaggle 
#https://www.kaggle.com/datasets/ysthehurricane/tesla-stock-data-20162021
```

```python
###importing all the necessary libraries

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
###collecting the data 


data=pd.read_csv("C:\DATASETS\TESLA_stock_prediction.csv")

data
```

```python
###the dataset we have is from the 2016-08-16 to 2021-08-13 


###data preprocessing and allocation of the data
data=data.drop(["Date"],axis=1)
```

```python

y=data["Adj Close"]

x=data.drop(["Adj Close"],axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

```

```python
x_train,x_test
```

```python
y_train,y_test
```

```python
####model building and fitting to the model 



from sklearn.ensemble import RandomForestRegressor


mod1=RandomForestRegressor()


mod1.fit(x_train,y_train)
```

```python
###model evaluation 






mod1.score(x_test,y_test)
```

```python
###making predictions 

result1=mod1.predict(x_test)


plt.plot(result1,color="red",label="predicted")
```

```python
fig,ax1=plt.subplots(1,1,figsize=(10,10))


ax1.bar(result1,y_test,color=("red","blue"))
ax1.plot(result1,y_test,color=("green"))
plt.xlabel("predicted vs Actual values")
```

```python
import pickle 

pickle.dump(mod1,open("model.pkl","wb"))
pickle.load(open("model.pkl","rb"))
            
```
